GRAPHS

```python
#8)comparision of states/union territories having highest and lowest total vaccinations dose?
plt.figure(figsize=(20, 10))
sns.barplot(x='State/UTs', y='Total Vaccination Doses', data=df, palette='viridis')
plt.title('Total COVID-19 Vaccinations by State (India)', fontsize=16, fontweight='bold')
plt.xlabel('State/UTs',fontsize=16, fontweight='bold')
plt.xticks(rotation=90)
plt.ylabel('Total Vaccination Doses',fontsize=14, fontweight='bold')
plt.tight_layout()
plt.show()
```

```python
#9)comparision of states having highest and lowest vaccinations of dose1 ?
plt.figure(figsize=(20, 10))
sns.barplot(x='State/UTs', y='Dose1', data=df,palette='pastel')
plt.title('Total COVID-19 Vaccinations Dose 1 by State (India)', fontsize=16, fontweight='bold')
plt.xlabel('State/UTs',fontsize=16, fontweight='bold')
plt.xticks(rotation=90)
plt.ylabel('Dose1',fontsize=14, fontweight='bold')
plt.tight_layout()
plt.show()
```

```python
#10)comparision of states having highest and lowest vaccinations of dose 2 ?
plt.figure(figsize=(20, 10))
sns.barplot(x='State/UTs', y='Dose 2', data=df, palette='pastel')
plt.title('Total COVID-19 Vaccinations Dose 2 by State (India)', fontsize=16, fontweight='bold')
plt.xlabel('State/UTs',fontsize=16, fontweight='bold')
plt.xticks(rotation=90)
plt.ylabel('Dose 2',fontsize=14, fontweight='bold')
plt.tight_layout()
plt.show()
```

```python
#11)comparison of total population and people treated with vaccination dose 1 ?
grouped = df.groupby('State/UTs')[['Dose1', 'Population']].sum().reset_index()
#grouped = grouped.sort_values(by='Population', ascending=False)
x = np.arange(len(grouped['State/UTs']))
width = 0.4
plt.figure(figsize=(16, 8))
plt.bar(x - width/2, grouped['Population'], width, label='Total Population', color='skyblue')
plt.bar(x + width/2, grouped['Dose1'], width, label='First Dose Administered', color='green')
plt.xticks(x, grouped['State/UTs'], rotation=90)
plt.ylabel('Count')
plt.title('Comparison of Total Population vs First Dose Administered by State')
plt.legend()
plt.tight_layout()
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
#12)comparison of total population and people treated with vaccination dose 2 ?
grouped = df.groupby('State/UTs')[['Dose 2', 'Population']].sum().reset_index()
#grouped = grouped.sort_values(by='Population', ascending=False)
x = np.arange(len(grouped['State/UTs']))
width = 0.4
plt.figure(figsize=(16, 8))
plt.bar(x - width/2, grouped['Population'], width, label='Total Population', color='blue')
plt.bar(x + width/2, grouped['Dose 2'], width, label='Second Dose Administered', color='red')
plt.xticks(x, grouped['State/UTs'], rotation=90)
plt.ylabel('Count')
plt.title('Comparison of Total Population vs Second Dose Administered by State')
plt.legend()
plt.tight_layout()
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
#13)Comparsion of states by Vaccination Dose 1 of age grp 15 - 18
plt.figure(figsize=(20, 10))
sns.barplot(x='State/UTs', y='Dose 1 15-18', data=df)
plt.title('Total COVID-19 Vaccinations Dose 1 - Age 15-18yrs by State (India)', fontsize=16, fontweight='bold')
plt.xlabel('State/UTs',fontsize=16, fontweight='bold')
plt.xticks(rotation=90)
plt.ylabel('Dose 1 15-18',fontsize=14, fontweight='bold')
plt.tight_layout()
plt.show()
```

```python
#14)Comparsion of states by Vaccination Dose 1 of age grp 12 - 14
plt.figure(figsize=(20, 10))
sns.barplot(x='State/UTs', y='Dose 1 12-14', data=df)
plt.title('Total COVID-19 Vaccinations Dose 1 - Age 12-14yrs by State (India)', fontsize=16, fontweight='bold')
plt.xlabel('State/UTs',fontsize=16, fontweight='bold')
plt.xticks(rotation=90)
plt.ylabel('Dose 1 12-14',fontsize=14, fontweight='bold')
plt.tight_layout()
plt.show()
```

```python
#15)Comparsion of states by Vaccination Dose 2 of age grp 15 - 18
plt.figure(figsize=(20, 10))
sns.barplot(x='State/UTs', y='Dose 2 15-18', data=df)
plt.title('Total COVID-19 Vaccinations Dose 2 - Age 15-18yrs by State (India)', fontsize=16, fontweight='bold')
plt.xlabel('State/UTs',fontsize=16, fontweight='bold')
plt.xticks(rotation=90)
plt.ylabel('Dose 2 15-18',fontsize=14, fontweight='bold')
plt.tight_layout()
plt.show()
```

```python
#16)Comparsion of states by Vaccination Dose 2 of age grp 12 - 14
plt.figure(figsize=(20, 10))
sns.barplot(x='State/UTs', y='Dose 2 12-14', data=df)
plt.title('Total COVID-19 Vaccinations Dose 2 - Age 12-14yrs by State (India)', fontsize=16, fontweight='bold')
plt.xlabel('State/UTs',fontsize=16, fontweight='bold')
plt.xticks(rotation=90)
plt.ylabel('Dose 2 12-14',fontsize=14, fontweight='bold')
plt.tight_layout()
plt.show()
```

```python
#17)Comparison of dose 1 and dose 2 of age grp 15 - 18
grouped = df.groupby('State/UTs')[['Dose 1 15-18', 'Dose 2 15-18']].sum().reset_index()
#grouped = grouped.sort_values(by='Dose 2 15-18', ascending=False)
x = np.arange(len(grouped['State/UTs']))
width = 0.4
plt.figure(figsize=(16, 8))
plt.bar(x + width/2, grouped['Dose 1 15-18'], width, label='Dose 1', color='red')
plt.bar(x - width/2, grouped['Dose 2 15-18'], width, label='Dose 2', color='blue')

plt.xticks(x, grouped['State/UTs'], rotation=90)
plt.ylabel('Count')
plt.title(' Dose 1 vs Dose 2 of age grp 15 - 18')
plt.legend()
plt.tight_layout()
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
#18)Comparison of dose 1 and dose 2 of age grp 12 - 14
grouped = df.groupby('State/UTs')[['Dose 1 12-14', 'Dose 2 12-14']].sum().reset_index()
#grouped = grouped.sort_values(by='Dose 2 12-14', ascending=False)
x = np.arange(len(grouped['State/UTs']))
width = 0.4
plt.figure(figsize=(16, 8))
plt.bar(x + width/2, grouped['Dose 1 12-14'], width, label='Dose 1', color='red')
plt.bar(x - width/2, grouped['Dose 2 12-14'], width, label='Dose 2', color='blue')

plt.xticks(x, grouped['State/UTs'], rotation=90)
plt.ylabel('Count')
plt.title(' Dose 1 vs Dose 2 of age grp 12 - 14')
plt.legend()
plt.tight_layout()
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

FEATURE ENGINEERING

```python
#20)dose 1 vs dose 2 coverage in %
df['Dose1_Coverage_%'] = (df['Dose1'] / df['Population']) * 100
df['Dose2_Coverage_%'] = (df['Dose 2'] / df['Population']) * 100
coverage_df = df[['State/UTs', 'Dose1_Coverage_%', 'Dose2_Coverage_%']]
coverage_df
```

```python
#21)Dose 1 Coverage (%) -graphical
df['Dose1_Coverage_%'] = (df['Dose1'] / df['Population']) * 100

plt.figure(figsize=(12, 10))
sns.barplot(data=df, x="Dose1_Coverage_%", y="State/UTs", color='skyblue')

plt.title("Dose 1 Percentage by State")
plt.xlabel("Coverage %")
plt.ylabel("State/UTs")
plt.legend()
plt.tight_layout()
plt.show()
```

```python
#22)Dose 2 Coverage (%) - graphical
df['Dose2_Coverage_%'] = (df['Dose 2'] / df['Population']) * 100

plt.figure(figsize=(12, 10))
sns.barplot(data=df, x="Dose2_Coverage_%", y="State/UTs", color='skyblue')

plt.title("Dose 2 Percentage by State")
plt.xlabel("Coverage %")
plt.ylabel("State/UTs")
plt.legend()
plt.tight_layout()
plt.show()
```

```python
#23)Dose 1 vs Dose 2 Coverage (%) - graphical
df['Dose1_Coverage_%'] = (df['Dose1'] / df['Population']) * 100
df['Dose2_Coverage_%'] = (df['Dose 2'] / df['Population']) * 100
df_sorted = df.sort_values("Dose1_Coverage_%", ascending=False)

plt.figure(figsize=(12, 10))
sns.barplot(data=df_sorted, x="Dose1_Coverage_%", y="State/UTs", color='skyblue', label='Dose 1')
sns.barplot(data=df_sorted, x="Dose2_Coverage_%", y="State/UTs", color='navy', alpha=0.6, label='Dose 2')

plt.title("Dose 1 and Dose 2 Coverage Percentage by State")
plt.xlabel("Coverage %")
plt.ylabel("State/UTs")
plt.legend()
plt.tight_layout()
plt.show()
```

```python
#24)Total teen dose coverage - age grp 12 to 18
df['Teen_Doses_Total'] = df['Dose 1 15-18'] + df['Dose 2 15-18'] + df['Dose 1 12-14'] + df['Dose 2 12-14']
teen_dose_df = df[['State/UTs', 'Teen_Doses_Total']]
teen_dose_df
```

```python
#25)Total dose administration to teenagers(12-18) - graphical
df['Teen_Doses_Total'] = df['Dose 1 15-18'] + df['Dose 2 15-18'] + df['Dose 1 12-14'] + df['Dose 2 12-14']
#df_sorted_teen = df.sort_values("Teen_Doses_Total", ascending=False)

plt.figure(figsize=(12, 10))
sns.barplot(data=df, x="Teen_Doses_Total", y="State/UTs")

plt.title("Total Doses Administered to Teenagers (1218)")
plt.xlabel("Teen Doses Total")
plt.ylabel("State/UTs")
plt.tight_layout()
plt.show()
```

```python
#26)Vaccination dose per person by state
df['Vaccination_per_Person'] = df['Total Vaccination Doses'] / df['Population']
vaccine_per_person_df = df[['State/UTs', 'Vaccination_per_Person']]
vaccine_per_person_df
```

```python
#27)graphical representation of Vaccination dose per person by state
df['Vaccination_per_Person'] = df['Total Vaccination Doses'] / df['Population']
#df_sorted_vpp = df.sort_values("Vaccination_per_Person", ascending=False)

plt.figure(figsize=(12, 10))
sns.barplot(data=df, x="Vaccination_per_Person", y="State/UTs", palette="viridis")

plt.title("Vaccination Doses per Person by State")
plt.xlabel("Doses per Person")
plt.ylabel("State/UTs")
plt.tight_layout()
plt.show()
```

```python
#28)Precaution dose coverage in %
df['Precaution_Coverage_%'] = (df['Precaution 18-59'] / df['Population']) * 100
precaution_df= df[['State/UTs', 'Precaution_Coverage_%']]
precaution_df
```

```python
#29)Graphical representation of Precaution dose coverage in %
df['Precaution_Coverage_%'] = (df['Precaution 18-59'] / df['Population']) * 100
plt.figure(figsize=(12, 10))
sns.barplot(data=df, x="Precaution_Coverage_%", y="State/UTs", palette="viridis")
plt.title("Precaution dose for age grp 18 - 59 by State")
plt.xlabel("Doses per Person")
plt.ylabel("State/UTs")
plt.tight_layout()
plt.show()
```

```python
#30)Estimation of unvaccinated people during vaccination dose 1 administation
df['Estimated_Unvaccinated_dose1'] = df['Population'] - df['Dose1']
df['unvaccinated_dose1%'] = (df['Estimated_Unvaccinated_dose1'] / df['Population']) * 100
unvaccinate_dose1_df = df[['State/UTs', 'Estimated_Unvaccinated_dose1','unvaccinated_dose1%']]
unvaccinate_dose1_df
```

```python
#31)Estimation of unvaccinated people during vaccination dose 2 administation
df['Estimated_Unvaccinated_dose2'] = df['Population'] - df['Dose 2']
df['unvaccinated_dose2%'] = (df['Estimated_Unvaccinated_dose2'] / df['Population']) * 100
unvaccinate_dose2_df = df[['State/UTs', 'Estimated_Unvaccinated_dose2','unvaccinated_dose2%']]
unvaccinate_dose2_df
```

```python
#32)Estimation of unvaccinated people during vaccination dose 1 vs dose 2 administation - graphical
grouped = df.groupby('State/UTs')[['unvaccinated_dose1%', 'unvaccinated_dose2%']].sum().reset_index()
#grouped = grouped.sort_values(by='Dose 2 12-14', ascending=False)
x = np.arange(len(grouped['State/UTs']))
width = 0.4
plt.figure(figsize=(16, 8))
plt.bar(x + width/2, grouped['unvaccinated_dose1%'], width, label='Dose 1', color='red')
plt.bar(x - width/2, grouped['unvaccinated_dose2%'], width, label='Dose 2', color='blue')

plt.xticks(x, grouped['State/UTs'], rotation=90)
plt.ylabel('Count %')
plt.title(' Unvaccinated dose 1 vs dose 2 in %')
plt.legend()
plt.tight_layout()
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

CORRELATION MATRIX

```python
#34)Create a correlation matrix between dose 1, dose 2, Dose 1 and 2 of age 15-18, Dose 1 and 2 of age 12-14 and precaution of age grp 18-59
corr_matrix = df[["Dose1" , "Dose 2" ,"Dose 1 15-18" ,"Dose 2 15-18" , "Dose 1 12-14", "Dose 2 12-14","Precaution 18-59"	]].corr()
plt.figure(figsize=(12, 8))
sns.heatmap(corr_matrix, annot=True, fmt=".2f", cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```
